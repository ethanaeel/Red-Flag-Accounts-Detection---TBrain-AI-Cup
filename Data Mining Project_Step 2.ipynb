Next we will flag each transaction with flag_x and flag_y

If event_date_x and event_date_y is null, flag_x and flag_y = 0

If txn_date is within D-2 to D of event_date_x, flag_x = 1

If txn_date is within D-2 to D of event_date_y, flag_y = 1


Otherwise = 0

```python
def flag_transaction(row):
    txn = row['txn_date']
    ex = row['event_date_x']
    ey = row['event_date_y']

    # Check sender flag
    flag_x = pd.notnull(ex) and (ex - 2 <= txn <= ex)
    
    # Check receiver flag
    flag_y = pd.notnull(ey) and (ey - 2 <= txn <= ey)
    
    return pd.Series({'flag_x': int(flag_x), 'flag_y': int(flag_y)})
```

```python
merged_df_all[['flag_x','flag_y']] = merged_df_all.apply(flag_transaction, axis=1)
```
